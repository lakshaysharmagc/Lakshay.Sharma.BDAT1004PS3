```python
females=users.where(users.gender=='F').groupby(['occupation','gender']).gender.agg(['count']) #step 8 - the percentage of women and men each occupation
males=users.where(users.gender=='M').groupby(['occupation','gender']).gender.agg(['count'])
total=users.groupby('occupation').gender.agg(['count'])
male_percent=(males/total)*100
female_percent=(females/total)*100
pd.merge(male_percent,female_percent,on='occupation')
```

## Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!


```python
import matplotlib.pyplot as plt

marriage=pd.read_csv('us-marriages-divorces-1867-2014.csv') # reading the data
year=marriage.Year.values 
marry=marriage.Marriages_per_1000.values
div=marriage.Divorces_per_1000.values

plt.plot(year,marry,color="Yellow" ,label="Marriage")
plt.plot(year,div,color="Red", label="Divorce")
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.ylabel("Marriages(Yellow) and divoces (red) per capita")
plt.title("marriages and divorces per capita in the U.S. between 1867 and 2014. \n\n\n")
plt.show()
```

## Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.

```python
y1=marriage.Year>=1900
y2=marriage.Year<=2000
data=marriage[y1 & y2]
plt.bar(data['Year'],data['Marriages_per_1000'],color="Cyan", label="Marriage",width = 1)
plt.bar(data['Year'],data['Divorces_per_1000'],color="grey", label="Divorce",width =1)
plt.title("Marraiges and Divorces per capita in US between 1900 to 2000 \n\n\n")
plt.xlabel("Years from 1900 to 2000")
plt.legend(loc="upper right")
plt.ylabel("Marriages & Divorces per capita in US")
plt.show()
```

## Question 9
Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.


```python
romanData= pd.read_csv('roman-emperor-reigns.csv')
romanData.head()
lowerData= romanData.where(romanData.Cause_of_Death=="Assassinated").Cause_of_Death.count()
remaining= romanData.Cause_of_Death.count()-lowerData
label=["Other Cause of Deaths","Assassinated"]
plt.pie([remaining,lowerData],labels=label,autopct='%.2f%%',shadow = True , radius = 2, colors= ['Red','Green','Yellow'])
plt.title("Roman Emperors \n\n")
plt.show()
```
